```python
data = pd.read_csv("actual.csv")
#data.head(5)
# 13,329

# Remove rows with missing data
data.dropna(inplace=True)
# 5,162

# Remove before 1996
data.drop(index=data[data['Year']<1996].index, inplace=True)
# 3,743

# Remove non-African countries
data.drop(index=data[data['Country'].isin(["Yugoslavia", "Afghanistan", "Qatar", "Bolivia", "Indonesia", "Suriname",
"Andorra", "Canada", "Pakistan", "Belize", "Micronesia, Fed. Sts.", "Monaco", "Kyrgyz Republic", "Iran, Islamic Rep.",
"Nepal", "Kuwait", "Peru", "Ecuador", "Trinidad and Tobago", "Thailand", "Bosnia and Herzegovina",
"United Arab Emirates", "Guyana", "Kazakhstan", "Bhutan", "Colombia", "Jordan", "Cuba", "Malaysia", "Latvia", "Nauru",
"Serbia", "Liechtenstein", "Belgium", "Moldova", "Panama", "Fiji", "Mexico", "Brunei Darussalam", "Brazil",
"Syrian Arab Republic", "Switzerland", "Luxembourg", "Lao PDR", "Guatemala", "Tajikistan", "Estonia", "Myanmar",
"North Macedonia", "Bahrain", "Venezuela, RB", "Georgia", "United States", "Nicaragua", "Ukraine", "Oman", "Palau",
"Dominican Republic", "Bahamas, The", "India", "Spain", "Sri Lanka", "Jamaica", "Uzbekistan", "Bulgaria", "New Zealand",
"Turkmenistan", "Singapore", "Croatia", "Iraq", "Mongolia", "Israel", "Lithuania", "Belarus", "Czech Republic",
"Turkey", "Romania", "St. Vincent and the Grenadines", "San Marino", "Taiwan", "Papua New Guinea", "Grenada",
"Argentina", "Slovak Republic", "Uruguay", "Russian Federation", "Philippines", "Vietnam", "Costa Rica", "Slovenia",
"Albania", "Cambodia", "Azerbaijan", "Dominica", "El Salvador", "Honduras", "Chile", "St. Kitts and Nevis",
"Saudi Arabia", "St. Lucia", "Paraguay", "Germany", "Antigua and Barbuda", "Tuvalu", "Greece", "China", "Hungary",
"Barbados", "Samoa", "Finland", "Lebanon", "Armenia", "United Kingdom", "Ireland", "Poland", "Italy", "Solomon Islands",
"Austria", "Netherlands", "France", "Haiti", "Cyprus", "Australia", "Tonga", "Denmark", "Iceland",
"Hong Kong SAR, China", "Marshall Islands", "Sweden", "Norway", "Kiribati", "Portugal", "Bangladesh", "Vanuatu",
"Malta", "Korea, Rep.", "Japan", "Korea, Dem. Peoples Rep.", "American Samoa", "Aruba", "Bermuda", "Timor-Leste",
"Faroe Islands", "French Guiana", "French Polynesia", "Gaza Strip", "Greenland", "Guadeloupe", "Guam", "Isle of Man",
"Jersey", "Macau", "Martinique", "Netherlands Antilles", "New Caledonia", "Northern Mariana Islands", "Puerto Rico",
"Virgin Islands (U.S.)", "West Bank and Gaza", "Yemen, Rep."])].index, inplace=True)
# 1,199

# Create country-year feature
countryyear = ["{}-{}".format(x[1]['Country'], x[1]['Year']) for x in data.iterrows()]
data["Country-Year"] = countryyear
data.set_index("Country-Year")
#data.head(5)

# Clarify Ongoing Conflict vs. New Outbreak features
data.rename(columns={"conflict":"Ongoing Conflict"}, inplace=True)
data.rename(columns={"emergence":"New Conflict"}, inplace=True)

# Creating Delta features

print(data.count()[0])
delta_debt=[np.nan for x in range(data.count()[0])]
delta_acct=[np.nan for x in range(data.count()[0])]
delta_corrupt=[np.nan for x in range(data.count()[0])]
delta_invest=[np.nan for x in range(data.count()[0])]
delta_gdp=[np.nan for x in range(data.count()[0])]
delta_infl=[np.nan for x in range(data.count()[0])]
delta_ethnic=[np.nan for x in range(data.count()[0])]
delta_relig=[np.nan for x in range(data.count()[0])]
delta_unemp=[np.nan for x in range(data.count()[0])]
delta_natresc=[np.nan for x in range(data.count()[0])]
delta_infmort=[np.nan for x in range(data.count()[0])]

index=1
for x in data.iterrows():
    for y in data.iterrows():
        if x[1]["Country"] == y[1]["Country"] and x[1]["Year"] == y[1]["Year"]-1:
            print("{}-{}, {}-{}".format(x[1]["Country"], x[1]["Year"], y[1]["Country"], y[1]["Year"]))
            delta_debt[index] = y[1]["Debt"]-x[1]["Debt"]
            delta_acct[index] = y[1]["Account Balance"]-x[1]["Account Balance"]
            delta_corrupt[index] = y[1]["Corruption"]-x[1]["Corruption"]
            delta_invest[index] = y[1]["Foreign_Investment"]-x[1]["Foreign_Investment"]
            delta_gdp[index] = y[1]["GDP"]-x[1]["GDP"]
            delta_infl[index] = y[1]["Inflation"]-x[1]["Inflation"]
            delta_ethnic[index] = y[1]["Ethnicity"]-x[1]["Ethnicity"]
            delta_relig[index] = y[1]["Religion"]-x[1]["Religion"]
            delta_unemp[index] = y[1]["unemployment"]-x[1]["unemployment"]
            delta_natresc[index] = y[1]["Natural Ressources"]-x[1]["Natural Ressources"]
            delta_infmort[index] = y[1]["Infant Mortality"]-x[1]["Infant Mortality"]
            break
    index+=1

data["Delta_Debt"] = delta_debt
data["Delta_Account_Balance"] = delta_acct
data["Delta_Corruption"] = delta_corrupt
data["Delta_Investment"] = delta_invest
data["Delta_GDP"] = delta_gdp
data["Delta_Inflation"] = delta_infl
data["Delta_Ethnicity"] = delta_ethnic
data["Delta_Religion"] = delta_relig
data["Delta_Unemployment"] = delta_unemp
data["Delta_Natural_Resources"] = delta_natresc
data["Delta_Infant_Mortality"] = delta_infmort

# Remove rows with missing data
data.dropna(inplace=True)
# 1,026

# Dropping unnecessary features
data.drop(columns=["Language", "GeoPoint", "incompatibility", "intensity_level", "minor conflict", "war",
                   "internationalized internal", "internal", "interstate", "extrasymetric", "War emergence"],
          inplace=True)

print(data.shape[0])
data.head(20)
```
