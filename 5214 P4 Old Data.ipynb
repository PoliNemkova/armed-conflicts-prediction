```python
import pandas as pd 
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics 
```

```python
data = '/Users/William/Desktop/Jupyter/P4Data.csv'
data = pd.read_csv(data)
```

```python
data.head()
```

```python
#X = data[['Debt', 'Account Balance', 'Corruption', 'Foreign_Investment', 'GDP', 'Inflation',]]
X = data.drop(['New Conflict', 'Country', 'Country Code', 'Country-Year', 'Delta_Corruption', 'Delta_Ethnicity', 'Delta_Religion'], axis = 1)
y = data['New Conflict']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
clf = RandomForestClassifier(n_estimators = 100)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```
